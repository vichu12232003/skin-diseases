```python
# Import necessary modules
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import models, transforms
from datasets import load_dataset
from torch.utils.data import DataLoader, Dataset
from PIL import Image
import os
```

```python
# Check GPU availability
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
# Load dataset from Hugging Face
print("Loading dataset...")
dataset = load_dataset("ahmed-ai/skin-lesions-classification-dataset")
print("Dataset loaded successfully!")
```

```python
# Custom dataset class
class SkinDataset(Dataset):
    def __init__(self, dataset, split="train"):
        self.data = dataset[split]
        self.transform = transforms.Compose([
            transforms.Resize((224, 224)),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        image = self.data[idx]["image"]
        if image.mode != "RGB":
            image = image.convert("RGB")
        image = self.transform(image)
        label = self.data[idx]["label"]
        return image, label
```

```python
# Create DataLoader (Optimized for GPU training)
print("Preparing DataLoader...")
train_dataset = SkinDataset(dataset)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True, num_workers=4, pin_memory=True)
print("DataLoader ready!")
```

```python
# Function to create a model
def create_model(model_name):
    if model_name == "resnet50":
        model = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
        model.fc = nn.Linear(model.fc.in_features, 14)
    elif model_name == "efficientnet_b3":
        model = models.efficientnet_b3(weights=models.EfficientNet_B3_Weights.DEFAULT)
        model.classifier[1] = nn.Linear(model.classifier[1].in_features, 14)
    elif model_name == "densenet121":
        model = models.densenet121(weights=models.DenseNet121_Weights.DEFAULT)
        model.classifier = nn.Linear(model.classifier.in_features, 14)
    elif model_name == "mobilenet_v3":  # Option to train a smaller model first
        model = models.mobilenet_v3_large(weights=models.MobileNet_V3_Large_Weights.DEFAULT)
        model.classifier[3] = nn.Linear(model.classifier[3].in_features, 14)
    return model
```

```python
# Training function (Fully optimized for GPU)
def train_model(model, train_loader, epochs=5):
    model.to(device)  # Move model to GPU
    criterion = nn.CrossEntropyLoss().to(device)  # Move loss function to GPU
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    print(f"Training started for {epochs} epochs...")

    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        for batch_idx, (images, labels) in enumerate(train_loader):
            images, labels = images.to(device), labels.to(device)  # Move data to GPU
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()

            if (batch_idx + 1) % 25 == 0:  # Print loss every 25 batches
                print(f"Epoch {epoch+1}/{epochs}, Batch {batch_idx+1}/{len(train_loader)}, Loss: {running_loss / (batch_idx+1):.4f}")

        print(f"Epoch {epoch+1} Completed! Loss: {running_loss/len(train_loader):.4f}")

    print("Training finished!")
    return model
```

```python
# Create directory for saving models
os.makedirs("trained_models", exist_ok=True)
```

```python
# Select models to train
models_to_train = ["resnet50", "efficientnet_b3", "densenet121"]
# To test a smaller model first, use: models_to_train = ["mobilenet_v3"]
```

```python
# Train and save models
for model_name in models_to_train:
    print(f"\n🚀 Training {model_name}...")
    model = create_model(model_name)
    trained_model = train_model(model, train_loader)
    torch.save(trained_model.state_dict(), f"trained_models/{model_name}.pth")
    print(f"✅ Saved {model_name}.pth")

# Monitor GPU usage while training
print("\nChecking GPU usage:")
!nvidia-smi
```
